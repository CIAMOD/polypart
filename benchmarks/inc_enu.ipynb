```python
import sys
from pathlib import Path
from time import perf_counter

import numpy as np

# Add the project root (parent of "nb") to sys.path
project_root = Path("..").resolve()
sys.path.insert(0, str(project_root))

from polypart.arrangements import get_moduli_arrangement, get_random_arrangement
from polypart.geometry import Hyperplane
from polypart.inc_enu import build_tree_inc_enu
from polypart.polytopes import (
    get_product_of_simplices,
    get_random_polytope_intersection,
)
```

```python
n, r = 1, 7
hyperplanes = get_moduli_arrangement(n, r, 0)
polytope = get_product_of_simplices(n, r - 1)

support = [Hyperplane(normal=a, offset=b) for a, b in zip(polytope.A, polytope.b)]

start_time = perf_counter()
root, n_partitions = build_tree_inc_enu(hyperplanes, support)
end_time = perf_counter()
print(f"Found {n_partitions} partitions in {end_time - start_time:.2f} seconds.")
```
