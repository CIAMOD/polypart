```python
from polypart.experiments import (
    ArrangementClass,
    Experiment,
    PolytopeClass,
    run_experiments,
    summarize_results,
)
```

```python
experiment = Experiment(
    P_class=PolytopeClass("random"),
    A_class=ArrangementClass("random", m=5, degen_ratio=1.0),
    d=2,
)
print(experiment)
results = experiment.run(decimals=3, seed=42)
print(results)
# experiment.save(results)
```

## Experiments


### Random Experiments Low Rank


```python
random_lowrank_experiments = [
    Experiment(
        P_class=PolytopeClass("random"),
        A_class=ArrangementClass("random", m=m, degen_ratio=0.0),
        d=d,
    )
    for d in [2, 3, 4]
    for m in [10, 20, 30, 40]
]

run_experiments(random_lowrank_experiments, n_runs=3, folder="../data")
```

```python
summarize_results(random_lowrank_experiments, folder="../data")
```

### Random Experiments


```python
random_experiments = [
    Experiment(
        P_class=PolytopeClass("random"),
        A_class=ArrangementClass("random", m=m, degen_ratio=degen_ratio),
        d=d,
    )
    for d in [2, 3, 4, 5, 6, 7]
    for m in [5, 10, 20]
    for degen_ratio in [0.0]
]

run_experiments(random_experiments, n_runs=5, folder="../data")
```

```python
summarize_results(random_experiments)
```

### Moduli Experiments


```python
moduli_experiments = [
    Experiment(
        P_class=PolytopeClass(moduli),
        A_class=ArrangementClass(moduli),
        d=d,
    )
    for moduli in ["moduli_n1", "moduli_r2"]
    for d in [2, 3, 4, 5, 6, 7]
    if not (moduli == "moduli_r2" and d > 6)
]

run_experiments(moduli_experiments, n_runs=3, folder="../data")
```

```python
summarize_results(moduli_experiments)
```
