```python
from polypart.experiments import (
    ArrangementClass,
    Experiment,
    PolytopeClass,
    plot_experiment_summary,
    run_experiments,
    summarize_results,
)
```

### Experiment: $d=3$, $m=30$ random polytope and arrangement


```python
d3m30 = Experiment(
    P_class=PolytopeClass("random"),
    A_class=ArrangementClass("random", m=30, degen_ratio=0.0),
    d=3,
)

run_experiments(d3m30, n_runs=10, folder="../data")
```

```python
plot_experiment_summary(d3m30)
```

```python
summarize_results(d3m30, folder="../data")
```
