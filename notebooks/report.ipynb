```python
from polypart.experiments import (
    ArrangementClass,
    Experiment,
    PolytopeClass,
    plot_experiment_summary,
    plot_times_per_m_across_dim,
    print_results_summary,
    run_experiments,
)
```

```python
experiments = [
    Experiment(
        PolytopeClass("random"),
        ArrangementClass("random", m=m, degen_ratio=0.0),
        d=d,
    )
    for d in [2, 3, 4, 5]
    for m in [10, 20, 30]
]
# run_experiments(experiments, n_runs=10, folder="../data")
plot_times_per_m_across_dim(experiments, folder="../data")
```

### Experiment: $d=3$, $m=30$ random polytope and arrangement


```python
d3m30 = Experiment(
    P_class=PolytopeClass("random"),
    A_class=ArrangementClass("random", m=30, degen_ratio=0.0),
    d=3,
)

# run_experiments(d3m30, n_runs=10, n_workers=3, folder="../data")
plot_experiment_summary(d3m30, folder="../data")
```

```python
print_results_summary(d3m30, folder="../data")
```
